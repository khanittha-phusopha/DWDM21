```python
path = '/content/drive/My Drive/DWDM_Project' #โดยเราจะทำการระบุให้ชี้ไปที่ โฟนเดอร์ที่เราเซฟไฟล์ไว้ หรือก็คือ โฟนเดอร์ DWDM_Project
```

```python
data_mf = pd.read_csv(os.path.join(path,'mf_port_th.csv')) #การนำเข้าไฟล์ที่มีตัวแปรคือdata_mf
data_mf
```

```python
data_pf = pd.read_csv(os.path.join(path,'pf_port_th.csv'))
data_pf
```

```python
data_pvd = pd.read_csv(os.path.join(path,'pvd_port_th.csv'))
data_pvd
```

```python
merged_data_mf_data_pf = data_mf.merge(data_pf,how='left',left_on='เงินลงทุนของกองทุนทั่วไปจำแนกตามประเภทสินทรัพย์',right_on='เงินลงทุนของกองทุนส่วนบุคคล จำแนกตามประเภทสินทรัพย์')
merged_data_mf_data_pf
```

```python
merged_total = data_pvd.merge(merged_data_mf_data_pf,how='left',left_on='เงินลงทุนของกองทุนสำรองเลี้ยงชีพ จำแนกตามประเภทสินทรัพย์',right_on='เงินลงทุนของกองทุนส่วนบุคคล จำแนกตามประเภทสินทรัพย์')
merged_total
```

```python
merged_total.dropna()
```
